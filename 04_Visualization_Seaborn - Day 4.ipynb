# Strip Plot

```python
# seaborn plots documentation - https://seaborn.pydata.org/tutorial/function_overview.html
# strip plot
sns.stripplot(y='tip',data=tips)
```

```python
# Groping with stripplot()
sns.stripplot(x='sex',y='tip',data=tips)
```

```python
sns.stripplot(x='day',y='tip',data=tips)
```

```python
sns.stripplot(x='day',y='tip',data=tips,size=3)
```

# Swarm Plot

```python
# Swarm Plot
sns.swarmplot(x='day',y='tip',data=tips,size=6)
```

```python
# high amount of tip is given by male and specifically on Saturday
sns.swarmplot(x='day',y='tip',data=tips,hue='sex',palette=['red','blue'])
# plt.ylabel=('tip($)')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip', data=tips, hue='time')
plt.show()
```

# Subplot

```python
# Subplot
plt.subplot(1,2,1) #nrows,ncols,position
sns.boxplot(x='day',y='tip',data=tips)

plt.subplot(1,2,2) #nrows,ncols,position
sns.boxplot(x='sex',y='tip',data=tips)
plt.tight_layout()

# plt.show()
```

# Scatterplot

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
# observation: tip amount is increasing with total bill value
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time')
# same points plotted but as per time - lunch & dinner
```

# Histplot

```python
sns.histplot(x='total_bill',data=tips)
```

```python
sns.histplot(x='tip',data=tips)
```

# Joint plot

```python
# Joint plot - Scatter plot and individual histogram of x and y variables

sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
pd.plotting.scatter_matrix(tips)
plt.tight_layout()
plt.show()
```

# Pair Plot

```python
#Pair plot : If dataset has many numeric columns, you can plot many plots of diff. combination using pairplot
# diagonally it gives histo or bar plot of individual plot of total_bill, tip and size
sns.pairplot(data=tips)
# sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex') ## diagonally male female distribution
plt.show()# try hue=time,smoker,day,etc. to check diff. patterns, gain more information
```
